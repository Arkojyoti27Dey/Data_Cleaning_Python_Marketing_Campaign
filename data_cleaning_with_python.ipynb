## Step-by-step (very simple)

### 1Ô∏è‚É£ `df.columns`

This gets **all column names** of your dataframe.

Example:

```text
[' Campaign ID ', 'Clicks (%)', 'Cost-per-Click$', 'Start Date (UTC)']
```

---

### 2Ô∏è‚É£ `.str.strip()`

üëâ Removes **spaces at the beginning and end**

```text
' Campaign ID ' ‚Üí 'Campaign ID'
```

---

### 3Ô∏è‚É£ `.str.lower()`

üëâ Converts everything to **lowercase**

```text
'Campaign ID' ‚Üí 'campaign id'
```

---

## ‚≠ê MOST IMPORTANT PART ‚≠ê

### 4Ô∏è‚É£ `.str.replace(r'[^a-z0-9]+', '_', regex=True)`

This is where the **magic** happens.

### Let‚Äôs break it into 3 pieces

#### üîπ A) `.replace( , )`

Means:

> ‚ÄúFind something and replace it with something else‚Äù

Here:

```python
.replace( PATTERN , '_' )
```

‚û° Replace the matched pattern with `_`

---

#### üîπ B) What is this pattern?

```python
r'[^a-z0-9]+'
```

### Read it slowly:

| Part  | Meaning     |
| ----- | ----------- |
| `a-z` | letters     |
| `0-9` | numbers     |
| `^`   | NOT         |
| `[]`  | group       |
| `+`   | one or more |

### üß† Plain English:

> **‚ÄúAnything that is NOT a letter or number‚Äù**

That includes:

* spaces
* `%`
* `$`
* `-`
* `/`
* `()`

---

#### üîπ C) `regex=True`

This tells Python:

> ‚ÄúTreat the pattern as a **rule**, not plain text.‚Äù

Without this, Python would look for the **exact string**
`[^a-z0-9]+` (which is useless).

---

### üîÑ Example transformation

| Before             | After            |
| ------------------ | ---------------- |
| `campaign id`      | `campaign_id`    |
| `clicks (%)`       | `clicks`         |
| `cost-per-click$`  | `cost_per_click` |
| `start date (utc)` | `start_date_utc` |

üëâ Everything **non-letter/non-number** becomes `_`

---

### 5Ô∏è‚É£ `.str.strip('_')`

üëâ Removes `_` from **start or end** (cleanup step)

```text
'_clicks_' ‚Üí 'clicks'
```

---

## üéØ Final result

### Before

```text
[' Campaign ID ', 'Clicks (%)', 'Cost-per-Click$', 'Start Date (UTC)']
```

### After

```text
['campaign_id', 'clicks', 'cost_per_click', 'start_date_utc']
```

---

# üßπ Start Date Cleaning ‚Äî What We Did & What Finally Worked

## 1Ô∏è‚É£ The original problem

Your dataset had a `start_date` column with **mixed formats**, for example:

* `2023-10-30`
* `2023-11-07 00:00:00`
* `22/02/2023`
* `21/08/2023`

Because of this, a **simple `pd.to_datetime()`** caused **hundreds of `NaT` values**, even though the dates were valid.

---

## 2Ô∏è‚É£ First attempts (why they didn‚Äôt fully work)

### ‚ùå Single-pass `pd.to_datetime(..., dayfirst=True)`

* Helped with `DD/MM/YYYY`
* Broke or conflicted with ISO formats (`YYYY-MM-DD`)
* Result: many `NaT`

---

### ‚ùå Multiple parsing passes on the same column

* Mixed strings, datetimes, and `NaT`
* Column stayed as `object`
* Pandas could not reliably infer dtype

---

### ‚ùå Regex-based selective parsing (ISO + slash)

* Technically correct
* Introduced complexity:

  * boolean mask issues (`NaN` in masks)
  * partial conversions
  * dtype inconsistencies
* Became fragile and hard to audit

---

### ‚ùå Overwriting the original column

* Made debugging difficult
* Hid which values were failing and why

---

## 3Ô∏è‚É£ Key realization (turning point)

> **The issue was not bad data ‚Äî it was the parsing strategy.**

* Dates were valid
* Pandas needs **row-wise format detection**, not a single global rule
* Re-parsing partially converted columns causes chaos

---

## 4Ô∏è‚É£ The correct professional approach we switched to

### ‚úÖ Preserve raw data

We **did not touch** the original columns:

* `start_date`
* `end_date`

Instead, we created:

* `start_date_new`
* `end_date_new`

This allowed:

* full auditability
* side-by-side comparison
* zero data loss

---

## 5Ô∏è‚É£ What finally worked (the winning solution)

### ‚úÖ Normalize first

* Convert values to clean strings
* Strip spaces
* Replace placeholders

### ‚úÖ Use **one deterministic conversion**

```python
pd.to_datetime(column, errors='coerce', format='mixed')
```

### Why this worked:

* Pandas parses **each row independently**
* Handles:

  * `YYYY-MM-DD`
  * `YYYY-MM-DD HH:MM:SS`
  * `DD/MM/YYYY`
* No regex
* No day-first conflicts
* No deprecated arguments
* No dtype corruption

---

## 6Ô∏è‚É£ Final result (success)

```text
start_date_new    0 NaT
end_date_new      0 NaT
```

```text
Both columns ‚Üí datetime64[ns]
```

‚úî All valid dates parsed
‚úî No data lost
‚úî Clean, auditable pipeline

---

## 7Ô∏è‚É£ The single most important lesson (remember this)

> **For mixed-format date columns, always normalize to strings and use `format='mixed'` in one pass.**

This avoids:

* multiple parsing passes
* mask errors
* dtype chaos
* silent failures

````python
# =====================================================
# ASSUMPTION
# start_date and end_date are CLEAN datetime64[ns]
# =====================================================

# -----------------------------------------------------
# STEP 1: Identify logically valid and invalid rows
# -----------------------------------------------------
valid_mask = marketing_campaign_df['end_date'] > marketing_campaign_df['start_date']
invalid_mask = ~valid_mask

print("Valid rows count:", valid_mask.sum())
print("Invalid rows count:", invalid_mask.sum())

# -----------------------------------------------------
# STEP 2: Calculate campaign duration (days) for valid rows
# -----------------------------------------------------
marketing_campaign_df.loc[valid_mask, 'duration_days'] = (
    marketing_campaign_df.loc[valid_mask, 'end_date']
    - marketing_campaign_df.loc[valid_mask, 'start_date']
).dt.days

print("\nSample valid durations:")
display(
    marketing_campaign_df.loc[
        valid_mask,
        ['start_date', 'end_date', 'duration_days']
    ].head(10)
)

# -----------------------------------------------------
# STEP 3: Remove outliers using IQR
# -----------------------------------------------------
q1 = marketing_campaign_df.loc[valid_mask, 'duration_days'].quantile(0.25)
q3 = marketing_campaign_df.loc[valid_mask, 'duration_days'].quantile(0.75)
iqr = q3 - q1

lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

print("\nIQR bounds for duration:")
print("Lower bound:", lower_bound)
print("Upper bound:", upper_bound)

non_outlier_mask = (
    valid_mask &
    (marketing_campaign_df['duration_days'] >= lower_bound) &
    (marketing_campaign_df['duration_days'] <= upper_bound)
)

print("Valid non-outlier rows:", non_outlier_mask.sum())

# -----------------------------------------------------
# STEP 4: Calculate average duration (outlier-free)
# -----------------------------------------------------
average_duration_days = marketing_campaign_df.loc[
    non_outlier_mask, 'duration_days'
].mean()

# üëâ IMPORTANT: round to integer days to avoid time component
avg_duration_days_int = int(round(average_duration_days))

print("\nAverage campaign duration (outlier-free):")
print(avg_duration_days_int, "days")

# -----------------------------------------------------
# STEP 5: Repair logically incorrect rows
# -----------------------------------------------------
marketing_campaign_df['timeline_repaired'] = False

marketing_campaign_df.loc[invalid_mask, 'duration_days'] = avg_duration_days_int

marketing_campaign_df.loc[invalid_mask, 'end_date'] = (
    marketing_campaign_df.loc[invalid_mask, 'start_date']
    + pd.to_timedelta(avg_duration_days_int, unit='D')
)

marketing_campaign_df.loc[invalid_mask, 'timeline_repaired'] = True

print("\nSample repaired rows:")
display(
    marketing_campaign_df.loc[
        invalid_mask,
        ['start_date', 'end_date', 'duration_days', 'timeline_repaired']
    ].head()
)

# -----------------------------------------------------
# STEP 6: Final validation
# -----------------------------------------------------
remaining_invalid = (
    marketing_campaign_df['end_date'] <= marketing_campaign_df['start_date']
).sum()

print("\nRemaining invalid timelines after repair:", remaining_invalid)

print("\nTimeline repaired flag distribution:")
print(marketing_campaign_df['timeline_repaired'].value_counts())
````

## ‚ùì Why can **spend be negative**?

Negative spend usually means **money was given back**, not money spent.

### The most common reasons:

1. **Refunds**

   * Ads were shown incorrectly
   * Invalid / bot clicks
   * Campaign stopped early
     ‚ûú Platform returns money ‚Üí shown as **negative spend**

2. **Billing corrections**

   * Platform overcharged earlier
   * Adjustment made later
     ‚ûú Correction appears as a **negative number**

3. **Credits**

   * Promotional or compensation credits
     ‚ûú Logged as negative spend

---

## ‚ùå What it does NOT mean

* It does **not** mean your campaign ‚Äúspent negative money‚Äù
* It is **not a data error**
* It is **not a calculation bug**

---

## üß† How should YOU treat it?

* **Keep negative spend** if you care about finance accuracy
* Optionally **flag it** so you know it‚Äôs an adjustment

```python
marketing_campaign_df['is_spend_adjustment'] = marketing_campaign_df['spend'] < 0
```

> **Negative spend = refund or correction, not performance.**
